```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
```

```python
data_frame=pd.read_csv("housepricedata.csv")
```

```python
data_frame.head()
```

```python
X=data_frame.iloc[:,0:10]
```

```python
y=data_frame.iloc[:,10]
```

```python
scaler=MinMaxScaler()
x_scale=scaler.fit_transform(X)
x_train,x_test,y_train,y_test=train_test_split(x_scale,y,test_size=0.3)
```

```python
model=Sequential([Dense(32,activation='relu',input_shape=(10,)),Dense(32,activation='relu'),Dense(1,activation='sigmoid'),])
```

```python
model.compile(optimizer='sgd',loss='binary_crossentropy')
```

```python
hist=model.fit(x_train,y_train,batch_size=300,epochs=100)
```

```python
acc=model.evaluate(x_test,y_test)
```

```python
print('Test Accuracy'+str((acc)))
```
